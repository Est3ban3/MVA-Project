```python
import ROOT
```

```python
#combine the two signal trees
sig_chain_db = ROOT.TChain("AnalysisMiniTree")
sig_chain_db.Add("PPSSP_2026/1l2tau/run2/signal_ggF.root")
sig_chain_db.Add("PPSSP_2026/1l2tau/run2/signal_VBF.root")

#opens and grabs background
f_bkg_db = ROOT.TFile("PPSSP_2026/1l2tau/run2/diboson.root")
t_bkg_db = f_bkg_db.Get("AnalysisMiniTree")

#call RDataFrame as its better for cuts
df_sig_db = ROOT.RDataFrame(sig_chain_db)
df_bkg_db = ROOT.RDataFrame(t_bkg_db)

#apply cuts
df_sig_db = df_sig_db.Filter("n_b_jet == 0 && n_jet >= 2")
df_bkg_db = df_bkg_db.Filter("n_b_jet == 0 && n_jet >= 2")

#convert met_met from MeV to GeV
df_sig_db = df_sig_db.Define("met_met_GeV", "met_met / 1000.")
df_bkg_db = df_bkg_db.Define("met_met_GeV", "met_met / 1000.")

#build histo
h_sig_db = df_sig_db.Histo1D(("h_sig_db", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")
h_bkg_db = df_bkg_db.Histo1D(("h_bkg_db", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")

#fill before normalising
h_sig_db.GetValue()
h_bkg_db.GetValue()

#normalises to 1
h_sig_db.Scale(1.0 / h_sig_db.Integral())
h_bkg_db.Scale(1.0 / h_bkg_db.Integral())

c_db = ROOT.TCanvas()
h_bkg_db.SetLineColor(ROOT.kPink + 2)
h_bkg_db.Draw("hist")
h_sig_db.Draw("hist same")

ROOT.gStyle.SetOptStat(0)

leg_db = ROOT.TLegend(0.62, 0.75, 0.88, 0.88)
leg_db.AddEntry(h_sig_db.GetPtr(), "Signal", "l")
leg_db.AddEntry(h_bkg_db.GetPtr(), "Diboson", "l")
leg_db.Draw()

c_db.SaveAs("met_sig_vs_bkg_db.png")
```

```python
#Combine Signal
sig_chain_sglH = ROOT.TChain("AnalysisMiniTree")
sig_chain_sglH.Add("PPSSP_2026/1l2tau/run2/signal_VBF.root")
sig_chain_sglH.Add("PPSSP_2026/1l2tau/run2/signal_ggF.root")

#Open and get background
f_sglH = ROOT.TFile("PPSSP_2026/1l2tau/run2/SingleH.root")
t_sglH = f_sglH.Get("AnalysisMiniTree")

#Call RDataFrame
df_signal_sglH = ROOT.RDataFrame(sig_chain_sglH)
df_bkg_sglH = ROOT.RDataFrame(t_sglH)

#Apply cuts
df_signal_sglH = df_signal_sglH.Filter("n_b_jet == 0 && n_jet >= 2")
df_bkg_sglH = df_bkg_sglH.Filter("n_b_jet == 0 && n_jet >= 2")

#Change to GeV
df_signal_sglH = df_signal_sglH.Define("met_met_GeV", "met_met / 1000")
df_bkg_sglH = df_bkg_sglH.Define("met_met_GeV", "met_met / 1000")

#build histo
h_sig_sglH = df_signal_sglH.Histo1D(("h_sig_sglH", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")
h_bkg_sglH = df_bkg_sglH.Histo1D(("h_bkg_sglH", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")

#fill
h_sig_sglH.GetValue()
h_bkg_sglH.GetValue()

#Normalise
h_sig_sglH.Scale(1.0 / h_sig_sglH.Integral())
h_bkg_sglH.Scale(1.0 / h_bkg_sglH.Integral())

#Draw
c_sglH = ROOT.TCanvas()
h_bkg_sglH.SetLineColor(ROOT.kPink + 2)
h_bkg_sglH.Draw("hist")
h_sig_sglH.Draw("hist same")

ROOT.gStyle.SetOptStat(0)

leg_sglH = ROOT.TLegend(0.62, 0.75, 0.88, 0.88)
leg_sglH.AddEntry(h_sig_sglH.GetPtr(), "Signal", "l")
leg_sglH.AddEntry(h_bkg_sglH.GetPtr(), "Single H", "l")
leg_sglH.Draw()

c_sglH.SaveAs("met_sig_vs_bkg_sglH.png")
```

```python
#Combine Signal
sig_chain_tops = ROOT.TChain("AnalysisMiniTree")
sig_chain_tops.Add("PPSSP_2026/1l2tau/run2/signal_VBF.root")
sig_chain_tops.Add("PPSSP_2026/1l2tau/run2/signal_ggF.root")

#Open and get background
f_tops = ROOT.TFile("PPSSP_2026/1l2tau/run2/tops.root")
t_tops = f_tops.Get("AnalysisMiniTree")

#Call RDataFrame
df_signal_tops = ROOT.RDataFrame(sig_chain_tops)
df_bkg_tops = ROOT.RDataFrame(t_tops)

#Apply cuts
df_signal_tops = df_signal_tops.Filter("n_b_jet == 0 && n_jet >= 2")
df_bkg_tops = df_bkg_tops.Filter("n_b_jet == 0 && n_jet >= 2")

#Change to GeV
df_signal_tops = df_signal_tops.Define("met_met_GeV", "met_met / 1000")
df_bkg_tops = df_bkg_tops.Define("met_met_GeV", "met_met / 1000")

#build histo
h_sig_tops = df_signal_tops.Histo1D(("h_sig_tops", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")
h_bkg_tops = df_bkg_tops.Histo1D(("h_bkg_tops", "Missing transverse energy;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")

#fill
h_sig_tops.GetValue()
h_bkg_tops.GetValue()

#Normalise
h_sig_tops.Scale(1.0 / h_sig_tops.Integral())
h_bkg_tops.Scale(1.0 / h_bkg_tops.Integral())

#set max
ymax_tops = max(h_sig_tops.GetMaximum(), h_bkg_tops.GetMaximum()) * 1.2
h_bkg_tops.SetMaximum(ymax_tops)

#Draw
c_tops = ROOT.TCanvas()
h_bkg_tops.SetLineColor(ROOT.kPink + 2)
h_bkg_tops.Draw("hist")
h_sig_tops.Draw("hist same")

ROOT.gStyle.SetOptStat(0)

leg_tops = ROOT.TLegend(0.62, 0.75, 0.88, 0.88)
leg_tops.AddEntry(h_sig_tops.GetPtr(), "Signal", "l")
leg_tops.AddEntry(h_bkg_tops.GetPtr(), "tops", "l")
leg_tops.Draw()

c_tops.SaveAs("met_sig_vs_bkg_tops.png")
```

```python
#Combine Signal
sig_chain_ttbar = ROOT.TChain("AnalysisMiniTree")
sig_chain_ttbar.Add("PPSSP_2026/1l2tau/run2/signal_VBF.root")
sig_chain_ttbar.Add("PPSSP_2026/1l2tau/run2/signal_ggF.root")

#Open and get background
f_ttbar = ROOT.TFile("PPSSP_2026/1l2tau/run2/ttbar.root")
t_ttbar = f_ttbar.Get("AnalysisMiniTree")

#Call RDataFrame
df_signal_ttbar = ROOT.RDataFrame(sig_chain_ttbar)
df_bkg_ttbar = ROOT.RDataFrame(t_ttbar)

#Apply cuts
df_signal_ttbar = df_signal_ttbar.Filter("n_b_jet == 0 && n_jet >= 2")
df_bkg_ttbar = df_bkg_ttbar.Filter("n_b_jet == 0 && n_jet >= 2")

#Change to GeV
df_signal_ttbar = df_signal_ttbar.Define("met_met_GeV", "met_met / 1000")
df_bkg_ttbar = df_bkg_ttbar.Define("met_met_GeV", "met_met / 1000")

#build histo
h_sig_ttbar = df_signal_ttbar.Histo1D(("h_sig_ttbar", "Missing transverse energy ;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")
h_bkg_ttbar = df_bkg_ttbar.Histo1D(("h_bkg_ttbar", "Missing transverse energy ;E_{T}^{miss} [GeV];Normalized to unit area", 50, 0, 200), "met_met_GeV", "weight")

#fill
h_sig_ttbar.GetValue()
h_bkg_ttbar.GetValue()

#Normalise
h_sig_ttbar.Scale(1.0 / h_sig_ttbar.Integral())
h_bkg_ttbar.Scale(1.0 / h_bkg_ttbar.Integral())

#Draw
c_ttbar = ROOT.TCanvas()
h_bkg_ttbar.SetLineColor(ROOT.kPink + 2)
h_bkg_ttbar.Draw("hist")
h_sig_ttbar.Draw("hist same")

ROOT.gStyle.SetOptStat(0)

leg_ttbar = ROOT.TLegend(0.62, 0.75, 0.88, 0.88)
leg_ttbar.AddEntry(h_sig_ttbar.GetPtr(), "Signal", "l")
leg_ttbar.AddEntry(h_bkg_ttbar.GetPtr(), "ttbar", "l")
leg_ttbar.Draw()

c_ttbar.SaveAs("met_sig_vs_bkg_ttbar.png")
```

```python
# Create a 2x2 canvas
c_mosaic = ROOT.TCanvas("c_mosaic", "All Met-Met Comparisons", 1200, 900)
c_mosaic.Divide(2, 2)

# Pad 1 (top-left): Diboson
c_mosaic.cd(1)
h_bkg_db.Draw("hist")
h_sig_db.Draw("hist same")
leg_db.Draw()

# Pad 2 (top-right): Single H
c_mosaic.cd(2)
h_bkg_sglH.Draw("hist")
h_sig_sglH.Draw("hist same")
leg_sglH.Draw()

# Pad 3 (bottom-left): Tops
c_mosaic.cd(3)
h_bkg_tops.Draw("hist")
h_sig_tops.Draw("hist same")
leg_tops.Draw()

# Pad 4 (bottom-right): ttbar
c_mosaic.cd(4)
h_bkg_ttbar.Draw("hist")
h_sig_ttbar.Draw("hist same")
leg_ttbar.Draw()

c_mosaic.SaveAs("met_all_backgrounds.png")
c_mosaic.Draw()
```
