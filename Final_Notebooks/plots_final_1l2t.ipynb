## Signal with all background for top 15 features

```python

import ROOT, math
from pathlib import Path

tree_name = "AnalysisMiniTree"
selection = "n_b_jet == 0 && n_jet >= 2"
base = Path("PPSSP_2026/1l2tau")
signal_files = {"signal_ggF.root", "signal_VBF.root"}

features = ["met_sumet", "tau1_pt", "dR_t1t2", "m_ttl", "tau2_pt",
            "m_t1t2", "dR_t1t2_l1", "HT_lepMET", "dR_l1j1", "l1_e",
            "m_l1j", "dR_l1t2", "dR_t1j1", "j2_pt", "m_all"]

periods = {
    "run2": [base / "run2"],
    "run3": [base / "run3"],
    "combined": [base / "run2", base / "run3"],
}

out_dir = Path("final plots 1l2t")
out_dir.mkdir(parents=True, exist_ok=True)

def paths(dirs, want_signal):
    return [str(p) for d in dirs for p in sorted(d.glob("*.root"))
            if (p.name in signal_files) == want_signal]

c = ROOT.TCanvas()
for pname, dirs in periods.items():
    sdf = ROOT.RDataFrame(tree_name, paths(dirs, True)).Filter(selection)
    bdf = ROOT.RDataFrame(tree_name, paths(dirs, False)).Filter(selection)
    ext = {f: (sdf.Min(f), sdf.Max(f), bdf.Min(f), bdf.Max(f)) for f in features}
    lims = {}
    for f, (smin, smax, bmin, bmax) in ext.items():
        lo = min(smin.GetValue(), bmin.GetValue())
        hi = max(smax.GetValue(), bmax.GetValue())
        if math.isfinite(lo) and math.isfinite(hi) and lo != hi:
            pad = (hi - lo) * 0.05
            lims[f] = (lo - pad, hi + pad)
    hists = {}
    for f, (lo, hi) in lims.items():
        hs = sdf.Histo1D((f"sig_{pname}_{f}", f"{f};{f};Normalised events", 50, lo, hi), f, "weights")
        hb = bdf.Histo1D((f"bkg_{pname}_{f}", f"{f};{f};Normalised events", 50, lo, hi), f, "weights")
        hists[f] = (hs, hb)
    for f, (hs, hb) in hists.items():
        s = hs.GetValue()
        b = hb.GetValue()
        if s.Integral() == 0 or b.Integral() == 0:
            continue
        s.Scale(1.0 / s.Integral())
        b.Scale(1.0 / b.Integral())
        s.SetLineColor(ROOT.kPink + 2)
        b.SetLineColor(ROOT.kAzure)
        b.SetMaximum(1.2 * max(s.GetMaximum(), b.GetMaximum()))
        b.Draw("hist")
        s.Draw("hist same")
        c.SaveAs(str(out_dir / f"1l2t_{pname}_{f}.png"))
    h.Draw("hist")
    c.SaveAs(str(out_dir / f"{Path(fname).stem}.png"))
```
